## **Menghitung Manual menggunakan Information gain untuk Klasifikasi**

### **Inisialisasi data berdasarkan kelas**

- Total data: 15 \\
- solanum lycopersicum (Tomat (0)): 8 data \\
- Solanum melongena (Terong (1)): 7 data

### **Menghitung Entropi Awal (S)**
$$
Entropi(D) = -p_1 \log_2(p_1) - p_2 \log_2(p_2)
$$

$$
p_1 = \frac{8}{15},\quad p_2 = \frac{7}{15}
$$

$$
Entropi(D) = -\frac{8}{15} \log_2 \frac{8}{15} - \frac{7}{15} \log_2 \frac{7}{15} \approx 0.996
$$

### **Fitur 1 (Tinggi Tanaman)**

### **Ambil Threshold**

  **Threshold**: 70 cm

  **Split**:
  - *Left* (≤70): No 6–14 → 7 data → semua kelas 1 (Terong)
  - *Right* (>70): No 1–11 → 8 data → semua kelas 0 (Tomat)

### **Hitung Entropi**

- **Entropi Left**
 - Semua data kelas 0(Tomat),probabilitas:
    $$
    p_{\text{Tomat}} = 0 , p_{\text{Terong}} = 1
    $$

 Maka:

  $$
  \text{Entropi(Left)} = - \left(1 \cdot \log_2 1 + 0 \cdot \log_2 0\right) = 0
  $$

- **Entropi Right**
 - Semua data kelas 0(Tomat),Probabilitas:
    $$
    p_{\text{Tomat}} = 1 , p_{\text{Terong}} = 0
    $$

 Maka:
  $$
  \text{Entropi(Right)} = - \left(1 \cdot \log_2 1 + 0 \cdot \log_2 0\right) = 0
  $$

- **Entropi Setelah Split**

  $$
  \text{Entropi}_{\text{split}} = \frac{7}{15} \cdot 0 + \frac{8}{15} \cdot 0 = 0
  $$

### **Hitung Information gain**
$$
IG = \text{Entropi}(S) - \left( \frac{7}{15} \cdot 0 + \frac{8}{15} \cdot 0 \right) = 0.996
$$

### **Hitung Gini Index**

- **Gini index sebelum split**

$$
p_0 = \frac{8}{15}, \quad p_1 = \frac{7}{15}
$$

$$
\text{Gini}(D) = 1 - \left( \frac{7}{15} \right)^2 - \left( \frac{8}{15} \right)^2
= 1 - \frac{49}{225} - \frac{64}{225}
= 1 - \frac{113}{225} \approx 0.498
$$

- **Gini index setelah split**
  - *Left*: semua kelas terong (Gini) : 0
  - *Right*: Semua kelas tomat (Gini) : 0

    $$
    \text{Gini}_{\text{split}} = \frac{7}{15} \cdot 0 + \frac{8}{15} \cdot 0 = 0
    $$

## **Visualisasi menggunakan DesicionTree**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.preprocessing import LabelEncoder
from matplotlib.colors import ListedColormap
from sklearn.tree import export_text

# === Data ===
data = {
    "Tinggi Tanaman (cm)": [75, 80, 78, 82, 79, 60, 62, 65, 63, 61, 85, 64, 77, 66, 76],
    "Diameter Daun (cm)": [11.2, 12.1, 10.9, 11.5, 11.8, 9.5, 9.7, 10.0, 9.8, 9.6, 12.3, 10.1, 11.0, 10.2, 11.1],
    "Kelas": [
        "Solanum lycopersicum", "Solanum lycopersicum", "Solanum lycopersicum", "Solanum lycopersicum", "Solanum lycopersicum",
        "Solanum melongena", "Solanum melongena", "Solanum melongena", "Solanum melongena", "Solanum melongena",
        "Solanum lycopersicum", "Solanum melongena", "Solanum lycopersicum", "Solanum melongena", "Solanum lycopersicum"
    ]
}
df = pd.DataFrame(data)

# === Label Encoding ===
le = LabelEncoder()
df['Label'] = le.fit_transform(df['Kelas'])  # lycopersicum=1, melongena=0 (tergantung urutan)

# === Fitur dan Target ===
X = df[["Tinggi Tanaman (cm)", "Diameter Daun (cm)"]]
y = df["Label"]  # gunakan label numerik hasil encoding

# === Train Decision Tree ===
clf = DecisionTreeClassifier(criterion="gini", max_depth=3, random_state=42)
clf.fit(X, y)

# === 1. Tampilkan Pohon Keputusan ===
print("\n=== Pohon Keputusan (Decision Tree) ===")
tree_rules = export_text(clf, feature_names=list(X.columns))
print(tree_rules)
```

## **Visualisasi Splitting Predict**

```python
# === Uji 2 Data Baru ===
uji_baru = pd.DataFrame({
    'Tinggi Tanaman (cm)': [67, 83],
    'Diameter Daun (cm)': [10.0, 11.9]
})

prediksi = clf.predict(uji_baru)
hasil = le.inverse_transform(prediksi)

print("\n=== Prediksi 2 Data Baru ===")
for i, (t, d) in enumerate(zip(uji_baru['Tinggi Tanaman (cm)'], uji_baru['Diameter Daun (cm)'])):
    print(f"Data {i+1}: Tinggi={t} cm, Diameter={d} cm → Prediksi: {hasil[i]}")

# === Visualisasi Decision Boundary ===
print("\n=== Visualisasi Decision Boundary ===")
plot_colors = "ry"
plot_step = 0.5

x_min, x_max = X['Tinggi Tanaman (cm)'].min() - 5, X['Tinggi Tanaman (cm)'].max() + 5
y_min, y_max = X['Diameter Daun (cm)'].min() - 1, X['Diameter Daun (cm)'].max() + 1

xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                     np.arange(y_min, y_max, plot_step))

grid = np.c_[xx.ravel(), yy.ravel()]
grid_df = pd.DataFrame(grid, columns=['Tinggi Tanaman (cm)', 'Diameter Daun (cm)'])
Z = clf.predict(grid_df)
Z = Z.reshape(xx.shape)

from matplotlib.colors import ListedColormap

# Buat colormap custom
custom_cmap = ListedColormap(["saddlebrown", "green"])

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap=custom_cmap, alpha=0.6)  # ganti dengan colormap coklat-hijau

# Plot titik data training
plot_colors = "yr"  # y = yellow, r = red
for label, color in zip([0, 1], plot_colors):
    idx = np.where(y == label)
    plt.scatter(X.iloc[idx]['Tinggi Tanaman (cm)'], X.iloc[idx]['Diameter Daun (cm)'],
                c=color, edgecolor='k', label=le.inverse_transform([label])[0])

# Plot data uji baru (tetap hitam)
plt.scatter(uji_baru['Tinggi Tanaman (cm)'], uji_baru['Diameter Daun (cm)'],
            c='black', marker='x', s=100, label='Data Uji Baru')

plt.xlabel("Tinggi Tanaman (cm)")
plt.ylabel("Diameter Daun (cm)")
plt.title("Decision Boundary (Tinggi vs Diameter Daun)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# === Data Latih ===
data = {
    "Tinggi Tanaman (cm)": [80, 75, 78, 82, 77, 60, 65, 63, 68, 66],
    "Diameter Daun (cm)": [11.5, 11.0, 11.2, 11.8, 11.3, 9.0, 8.5, 9.2, 8.8, 9.1],
    "Kelas": ["Tomat", "Tomat", "Tomat", "Tomat", "Tomat",
              "Terong", "Terong", "Terong", "Terong", "Terong"]
}
df = pd.DataFrame(data)

# === Label Encoding ===
le = LabelEncoder()
df['Label'] = le.fit_transform(df['Kelas'])  # Tomat=1, Terong=0

# === Fitur dan Target ===
X = df[["Tinggi Tanaman (cm)", "Diameter Daun (cm)"]]
y = df["Label"]

# === Train Decision Tree ===
clf = DecisionTreeClassifier(criterion="gini", max_depth=3, random_state=42)
clf.fit(X, y)

# === Data Uji Baru ===
uji_baru = pd.DataFrame({
    "Tinggi Tanaman (cm)": [67, 83],
    "Diameter Daun (cm)": [10.0, 11.9]
})

# === Visualisasi Decision Boundary ===
print("\n=== Visualisasi Decision Boundary ===")
plot_colors = "yr"  # y = kuning (Terong), r = merah (Tomat)
plot_step = 0.5

x_min, x_max = X['Tinggi Tanaman (cm)'].min() - 5, X['Tinggi Tanaman (cm)'].max() + 5
y_min, y_max = X['Diameter Daun (cm)'].min() - 1, X['Diameter Daun (cm)'].max() + 1

xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                     np.arange(y_min, y_max, plot_step))

grid = np.c_[xx.ravel(), yy.ravel()]
grid_df = pd.DataFrame(grid, columns=['Tinggi Tanaman (cm)', 'Diameter Daun (cm)'])
Z = clf.predict(grid_df)
Z = Z.reshape(xx.shape)

# Custom colormap untuk area (kelas)
custom_cmap = ListedColormap(["saddlebrown", "green"])  # Terong=coklat, Tomat=hijau

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap=custom_cmap, alpha=0.6)

# Plot titik data training
for label, color in zip([0, 1], plot_colors):
    idx = np.where(y == label)
    plt.scatter(X.iloc[idx]['Tinggi Tanaman (cm)'], X.iloc[idx]['Diameter Daun (cm)'],
                c=color, edgecolor='k', label=le.inverse_transform([label])[0])

# Plot data uji baru
plt.scatter(uji_baru['Tinggi Tanaman (cm)'], uji_baru['Diameter Daun (cm)'],
            c='black', marker='X', s=100, label='Data Uji Baru')

prediksi = clf.predict(uji_baru)
hasil = le.inverse_transform(prediksi)

for i, (t, d) in enumerate(zip(uji_baru['Tinggi Tanaman (cm)'], uji_baru['Diameter Daun (cm)'])):
    print(f"Data Uji {i+1}: Tinggi={t}, Diameter={d} → Prediksi: {hasil[i]}")

# Label sumbu dan tampilan
plt.xlabel("Tinggi Tanaman (cm)")
plt.ylabel("Diameter Daun (cm)")
plt.title("Decision Boundary (Tinggi vs Diameter Daun)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
